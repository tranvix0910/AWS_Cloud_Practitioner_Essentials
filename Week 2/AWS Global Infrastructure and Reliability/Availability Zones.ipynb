##### **Amazon EC2 instance in a single Availability Zone**

Gi s bn ang chy mt ng dng trn mt Amazon EC2 instances n l trong **Khu vc Bc California**. 

Instances ang chy trong **Khu vc Kh dng us-west-1a**. **Nu us-west-1a gp s c, bn s mt Instances** ca mnh.

##### **Amazon EC2 instances in multiple Availability Zones**

Mt thc hnh tt l **chy ng dng qua t nht hai Khu vc Kh dng** trong mt Khu vc. 

Trong v d ny, bn c th chn chy thm mt th hin Amazon EC2 th hai trong us-west-1b.
