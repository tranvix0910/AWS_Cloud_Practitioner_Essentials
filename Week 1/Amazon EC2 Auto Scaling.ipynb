## **Amazon EC2 Auto Scaling**

Cho php bn t ng **thm hoc loi b** cc Amazon EC2 instances  p ng yu cu thay i ca ng dng. 

Bng cch t ng co gin **(scaling in/out)** cc instances ca bn theo nhu cu, bn c th duy tr tnh sn c ca ng dng mt cch hiu qu hn.

Trong **Amazon EC2 Auto Scaling**, bn c th s dng hai phng php: 

- co gin ng (dynamic scaling) 

- co gin d on (predictive scaling).

**Co gin ng** 

- p ng yu cu thay i ca ng dng.

- Khi lu lng truy cp tng ln, **EC2 Auto Scaling** t ng thm cc intances  x l ti ln hn. 

- Ngc li, khi lu lng gim i, n s t ng gim s lng intances  tit kim ti nguyn.

**Co gin d on** 

- T ng ln lch cc intances Amazon EC2 ng s lng da trn d on yu cu. 

- iu ny gip bn chun b sn sng  p ng nhu cu d kin v trnh vic chm tr do khng c  ti nguyn.

Trong mi trng m my, sc mnh tnh ton l mt ngun ti nguyn c th lp trnh c, cho php bn tip cn mt cch linh hot vn  v **scaling**. 

Bng cch thm **Amazon EC2 Auto Scaling** vo ng dng, bn c th thm cc intances mi vo ng dng khi cn thit v chm dt chng khi khng cn na.

Khi bn to mt nhm **Auto Scaling**, bn c th **thit lp s lng ti thiu** ca cc intances Amazon EC2. 

**S lng ti thiu** 

- S lng cc intances Amazon EC2 c khi chy ngay sau khi bn to nhm Auto Scaling. 

- Trong v d ny, nhm Auto Scaling c mt s lng ti thiu l mt Amazon EC2 instances.

Tip theo, bn c th t s lng mong mun **(desired capacity)** l hai Amazon EC2 instances, ngay c khi ng dng ca bn ch cn ti thiu mt intances Amazon EC2  chy.

**Lu :** 

- Nu bn khng ch nh s lng mong mun ca cc intances Amazon EC2 trong mt nhm Auto Scaling, **s lng mong mun s mc nh l s lng ti thiu**.

Cu hnh th ba m bn c th t trong mt nhm Auto Scaling l **s lng ti a**. 

V d, bn c th cu hnh nhm **Auto Scaling**  **t ng co gin (scale out)**  p ng nhu cu tng ln, nhng ch ti mt **s lng ti a l bn Amazon EC2 instances**.

Bi v **Amazon EC2 Auto Scaling** s dng cc **Amazon EC2 instances**, bn ch phi tr tin cho cc intances bn s dng, khi bn s dng chng. 

By gi bn c mt kin trc tit kim chi ph cung cp tri nghim tt nht cho khch hng trong khi gim thiu chi ph.
