## **Amazon EC2 Auto Scaling**

Cho phép bạn tự động **thêm hoặc loại bỏ** các Amazon EC2 instances để đáp ứng yêu cầu thay đổi của ứng dụng. 

Bằng cách tự động co giãn **(scaling in/out)** các instances của bạn theo nhu cầu, bạn có thể duy trì tính sẵn có của ứng dụng một cách hiệu quả hơn.

Trong **Amazon EC2 Auto Scaling**, bạn có thể sử dụng hai phương pháp: 

- co giãn động (dynamic scaling) 

- co giãn dự đoán (predictive scaling).

**Co giãn động** 

- Đáp ứng yêu cầu thay đổi của ứng dụng.

- Khi lưu lượng truy cập tăng lên, **EC2 Auto Scaling** tự động thêm các intances để xử lý tải lớn hơn. 

- Ngược lại, khi lưu lượng giảm đi, nó sẽ tự động giảm số lượng intances để tiết kiệm tài nguyên.

**Co giãn dự đoán** 

- Tự động lên lịch các intances Amazon EC2 đúng số lượng dựa trên dự đoán yêu cầu. 

- Điều này giúp bạn chuẩn bị sẵn sàng để đáp ứng nhu cầu dự kiến và tránh việc chậm trễ do không có đủ tài nguyên.

Trong môi trường đám mây, sức mạnh tính toán là một nguồn tài nguyên có thể lập trình được, cho phép bạn tiếp cận một cách linh hoạt vấn đề về **scaling**. 

Bằng cách thêm **Amazon EC2 Auto Scaling** vào ứng dụng, bạn có thể thêm các intances mới vào ứng dụng khi cần thiết và chấm dứt chúng khi không cần nữa.

Khi bạn tạo một nhóm **Auto Scaling**, bạn có thể **thiết lập số lượng tối thiểu** của các intances Amazon EC2. 

**Số lượng tối thiểu** 

- Số lượng các intances Amazon EC2 được khởi chạy ngay sau khi bạn tạo nhóm Auto Scaling. 

- Trong ví dụ này, nhóm Auto Scaling có một số lượng tối thiểu là một Amazon EC2 instances.

Tiếp theo, bạn có thể đặt số lượng mong muốn **(desired capacity)** là hai Amazon EC2 instances, ngay cả khi ứng dụng của bạn chỉ cần tối thiểu một intances Amazon EC2 để chạy.

**Lưu ý:** 

- Nếu bạn không chỉ định số lượng mong muốn của các intances Amazon EC2 trong một nhóm Auto Scaling, **số lượng mong muốn sẽ mặc định là số lượng tối thiểu**.

Cấu hình thứ ba mà bạn có thể đặt trong một nhóm Auto Scaling là **số lượng tối đa**. 

Ví dụ, bạn có thể cấu hình nhóm **Auto Scaling** để **tự động co giãn (scale out)** để đáp ứng nhu cầu tăng lên, nhưng chỉ tới một **số lượng tối đa là bốn Amazon EC2 instances**.

Bởi vì **Amazon EC2 Auto Scaling** sử dụng các **Amazon EC2 instances**, bạn chỉ phải trả tiền cho các intances bạn sử dụng, khi bạn sử dụng chúng. 

Bây giờ bạn có một kiến trúc tiết kiệm chi phí cung cấp trải nghiệm tốt nhất cho khách hàng trong khi giảm thiểu chi phí.
