## **Amazon Simple Notification Service (Amazon SNS)**

Amazon Simple Notification Service (Amazon SNS) l mt dch v **publish/subscribe**.

S dng cc ch  **(topics)** trong **Amazon SNS**, ngi xut bn **(publisher)** gi thng ip ti cc ngi ng k **(subscribers)**.

Trong **Amazon SNS**, cc ngi ng k c th l cc my ch web, a ch email, cc hm AWS Lambda hoc mt s la chn khc.

##### **Publishing updates from a single topic ( Xut bn thng tin t mt ch  duy nht )**

Gi s qun c ph c mt bn tin duy nht bao gm cp nht t tt c cc lnh vc kinh doanh ca mnh. N bao gm cc ch  nh phiu gim gi, th v v c ph v sn phm mi. **Tt c cc ch  ny c nhm li v y l mt bn tin duy nht**. Tt c khch hng ng k nhn bn tin s nhn c cp nht v phiu gim gi, th v v c ph v sn phm mi.

Sau mt thi gian, mt s khch hng cho bit h mun nhn cc bn tin ring bit cho ch cc ch  c th m h quan tm. Ch ca hng quyt nh th tip cn ny.

#### **Publishing updates from multiple topics ( Xut bn thng tin t nhiu ch  )**

By gi, thay v c mt bn tin duy nht cho tt c cc ch , qun c ph  chia n thnh ba bn tin ring bit. 

Mi bn tin dnh ring cho mt ch  c th: phiu gim gi, th v v c ph v sn phm mi.

Ngi ng k s nhn ngay lp tc cp nht ch cho cc ch  c th m h  ng k.

C th ngi ng k ng k cho mt ch  duy nht hoc nhiu ch . 

V d, khch hng u tin ng k ch cho ch  phiu gim gi, v ngi ng k th hai ng k ch cho ch  th v v c ph. Khch hng th ba ng k c hai ch  th v v c ph v sn phm mi.
